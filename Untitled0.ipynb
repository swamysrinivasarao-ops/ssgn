```python
# ============================================
# Stress–Strain Analysis using Pandas
# ============================================

# Step 0: Import required libraries
import pandas as pd

# --------------------------------------------
# Step 1: Load the CSV file and display first 5 rows
# --------------------------------------------

# Load the CSV file (make sure the file is uploaded to Colab)
file_path = "stress_strain_practice.csv"
df = pd.read_csv(file_path)

# Display the first 5 rows
print("First 5 rows of the dataset:")
display(df.head())

# --------------------------------------------
# Step 2: Calculate Stress (MPa)
# Stress (MPa) = Force (N) / Area (mm²)
# Note: 1 N/mm² = 1 MPa
# --------------------------------------------

df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]

# --------------------------------------------
# Step 3: Calculate Strain (%)
# Strain (%) = (Elongation / Original Length) * 100
# --------------------------------------------

df["Strain_percent"] = (df["Elongation_mm"] / df["Original_Length_mm"]) * 100

# --------------------------------------------
# Step 4: Group by Material and calculate
# average Stress and average Strain
# --------------------------------------------

material_avg = (
    df.groupby("Material")[["Stress_MPa", "Strain_percent"]]
    .mean()
    .reset_index()
)

print("\nAverage Stress and Strain by Material:")
display(material_avg)

# --------------------------------------------
# Step 5: Identify the strongest material
# (highest average stress)
# --------------------------------------------

strongest_material = material_avg.loc[
    material_avg["Stress_MPa"].idxmax()
]

print("\nStrongest Material (Highest Average Stress):")
print(f"Material: {strongest_material['Material']}")
print(f"Average Stress (MPa): {strongest_material['Stress_MPa']:.2f}")

# --------------------------------------------
# Step 6: Summary statistics (min, max, average stress)
# for each material
# --------------------------------------------

stress_summary = (
    df.groupby("Material")["Stress_MPa"]
    .agg(
        Min_Stress_MPa="min",
        Max_Stress_MPa="max",
        Avg_Stress_MPa="mean"
    )
    .reset_index()
)

print("\nStress Summary by Material:")
display(stress_summary)

# --------------------------------------------
# Step 7: Export results to a new CSV file
# --------------------------------------------

output_file = "stress_strain_results_summary.csv"

# Merge average strain data with stress summary
final_results = pd.merge(
    stress_summary,
    material_avg,
    on="Material"
)

# Save to CSV
final_results.to_csv(output_file, index=False)

print(f"\nResults exported successfully to '{output_file}'")

# --------------------------------------------
# Optional: Display final exported table
# --------------------------------------------
display(final_results)
```

```python
# ============================================
# Stress–Strain Analysis with Visualization,
# Threshold Highlighting, and Error Handling
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# --------------------------------------------
# Step 1: Load CSV file with error handling
# --------------------------------------------
try:
    df = pd.read_csv("stress_strain_practice.csv")
except FileNotFoundError:
    raise FileNotFoundError(
        "File not found. Please upload 'stress_strain_practice.csv' before running this cell."
    )

print("First 5 rows of the dataset:")
display(df.head())

# --------------------------------------------
# Step 2: Handle missing values
# Drop rows with missing critical data
# --------------------------------------------

critical_columns = [
    "Material",
    "Force_N",
    "Cross_Section_Area_mm2",
    "Elongation_mm",
    "Original_Length_mm"
]

print("Missing values before cleaning:")
display(df[critical_columns].isnull().sum())

df = df.dropna(subset=critical_columns)

print("Missing values after cleaning:")
display(df[critical_columns].isnull().sum())

# --------------------------------------------
# Step 3: Calculate Stress and Strain
# --------------------------------------------

# Stress in MPa (1 N/mm² = 1 MPa)
df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]

# Strain in percentage
df["Strain_percent"] = (df["Elongation_mm"] / df["Original_Length_mm"]) * 100

# --------------------------------------------
# Step 4: Highlight tests where Stress > 200 MPa
# --------------------------------------------

high_stress_tests = df[df["Stress_MPa"] > 200]

print("Tests where Stress exceeded 200 MPa:")
display(high_stress_tests)

# --------------------------------------------
# Step 5: Group by Material and calculate averages
# --------------------------------------------

material_avg = (
    df.groupby("Material")[["Stress_MPa", "Strain_percent"]]
    .mean()
    .reset_index()
)

print("Average Stress and Strain by Material:")
display(material_avg)

# --------------------------------------------
# Step 6: Visualization
# Bar chart – Average Stress vs Material
# --------------------------------------------

plt.figure()
plt.bar(material_avg["Material"], material_avg["Stress_MPa"])
plt.xlabel("Material Type")
plt.ylabel("Average Stress (MPa)")
plt.title("Average Stress by Material")
plt.show()

# --------------------------------------------
# Step 7: Summary statistics for stress
# --------------------------------------------

stress_summary = (
    df.groupby("Material")["Stress_MPa"]
    .agg(
        Min_Stress_MPa="min",
        Max_Stress_MPa="max",
        Avg_Stress_MPa="mean"
    )
    .reset_index()
)

print("Stress Summary by Material:")
display(stress_summary)

# --------------------------------------------
# Step 8: Export final results to CSV
# --------------------------------------------

final_results = pd.merge(
    stress_summary,
    material_avg,
    on="Material"
)

final_results.to_csv("stress_strain_results_summary.csv", index=False)

print("Results successfully exported to 'stress_strain_results_summary.csv'")
```
